```python
import re
import json
import copy
```

```python
with(open("gameData_v05.json", "r") as fs):
  weapons = json.load(fs)
```

```python
def extract_hangul(text):
    # 괄호 안에 있는 문자열을 제외하고 한글만 추출하는 정규식입니다.
    hangul_regex = r'(?![^(]*\))([가-힣]+)'
    hangul_matches = re.findall(hangul_regex, text)
    
    # 추출된 한글을 합쳐서 반환합니다.
    if hangul_matches:
        return ''.join(hangul_matches)
    else:
        return '' # 한글이 없을 경우 빈 문자열 반환
```

```python
for weapon in weapons:
  attackTypes = []
  for damage in weapon['damage']:
    attackTypes.append(extract_hangul(damage))
  weapon["attackTypes"] = attackTypes
```
